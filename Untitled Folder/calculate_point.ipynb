```python
import matplotlib.pyplot as plt
import numpy as np

gatea = np.array([100, 100])
gateb = np.array([200, 200])
print(gatea)
```

```python
def calculate_access_point(gate1, gate2, offset, closer_point):
    difference = (gate1 - gate2) / 2
    middle = difference + gate2
    perpendicular_v = np.array([-difference[1], difference[0]])
    perpendicular_v /= np.linalg.norm(perpendicular_v)
    print(middle)
    point1 = middle + offset*perpendicular_v
    point2 = middle + offset*perpendicular_v*-1
    if np.linalg.norm(point1-closer_point) < np.linalg.norm(point2-closer_point):
        return point1
    else:
        return point2
```

```python
difference = (gateb - gatea) / 2
middle = difference + gatea
perpendicular_v = np.array([-difference[1], difference[0]])
perpendicular_v /= np.linalg.norm(perpendicular_v)
OFFSET = 100
```

```python
point1 = middle + OFFSET*perpendicular_v
point2 = middle + OFFSET*perpendicular_v*-1
```

```python
ROBOT = [70, 0]
circle1 = plt.Circle(gatea, 20, color='b')
circle2 = plt.Circle(gateb, 20, color='b')
robot = plt.Circle(ROBOT, 20, color='cyan')
point = plt.Circle(calculate_access_point(gatea, gateb, 100, ROBOT), 20, color='r')
circles = [circle1, circle2, robot, point]



fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
[ax.add_patch(patch) for patch in circles]
ax.set_xlim([-1000, 1000])
ax.set_ylim([-1000, 1000])
plt.show()
```
